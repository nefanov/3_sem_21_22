```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.arange(1, 9)
y1 = []
with open('result.txt', 'r') as file:
    y1 = [line.rstrip('\n') for line in file]
    y1 = [float(string) for string in y1]
```

```python
y2 = [y1[0]/(i + 1) for i in range(8)]
```

```python
plt.figure(figsize =(20, 10))
plt.grid()
plt.plot(x, y1, label='Real values')
plt.plot(x, y2, label='Values without sync time (Ideal)')
plt.legend(fontsize=25)
plt.show()
```
